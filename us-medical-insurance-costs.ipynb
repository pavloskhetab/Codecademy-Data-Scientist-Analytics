## Age Distribution
To view the age distribution we shall create a histogram.<br>
We will also calculate the mean, median, mode, variance and standard deviation to summarise the age distribution.

```python
import matplotlib.pyplot as plt
import numpy as np
import statistics

mean_age = statistics.mean(age)
median_age = statistics.median(age)
median_low_age = statistics.median_low(age)
median_high_age = statistics.median_high(age)
mode_age = statistics.multimode(age)
s_squared = statistics.variance(age)
s_age = statistics.stdev(age)

max_age = max(age)
min_age = min(age)

"""
Calculate the number of bins for a histogram using the Freedman-Diaconis rule.

Parameters:
data (list or numpy array): The data for which to calculate the number of bins.

Returns:
int: The number of bins.
"""
def calculate_num_bins(data):
    data_range = max(data) - min(data)  # Calculate the data range
    iqr = np.percentile(data, 75) - np.percentile(data, 25)  # Calculate the interquartile range
    bin_width = 2 * iqr / (len(data) ** (1/3))  # Calculate the bin width using Freedman-Diaconis rule
    num_bins = int(data_range / bin_width)  # Calculate the number of bins
    
    return num_bins, bin_width # Return both num_bins and bin_width

# Calculate the number of bins using the function
num_bins_age, bin_width_age = calculate_num_bins(age)

# Create the histogram with the specified number of bins and range
plt.hist(age, bins = num_bins_age, range = (min_age, max_age), edgecolor = 'k', alpha=0.8)

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')

# Create a list of additional tick positions you want to add
additional_ticks = [0, 5, 10, 15, 25, 35, 45, 55, 65]

# Get the current tick positions and labels
current_ticks = plt.xticks()[0].tolist()
current_labels = [int(label) for label in plt.xticks()[0]]

# Combine the current ticks with the additional ticks
new_ticks = current_ticks + additional_ticks
new_labels = current_labels + additional_ticks

# Set the new ticks and labels on the x-axis
plt.xticks(new_ticks, new_labels)

plt.show()

# Convert max_age and min_age to strings before printing
print('Max Age: ' + str(max_age))
print('Min Age: ' + str(min_age))
print(f'Bin Width for Age: {round(bin_width_age,2)}')
print(f'Mean Age: {mean_age}')
print(f'Median Age: {median_age}')
print(f'Median Low Age: {median_low_age}')
print(f'Median High Age: {median_high_age}')
print(f'Mode Age: {mode_age}')
print(f'Variance Age: {s_squared}')
print(f'Standard Deviation Age: {s_age}')
print(f'Num. of bins: {num_bins_age}')
```

## Understanding Q-Q Plots in Statistics: Assessing Data Distribution Fit
A Quantile-Quantile (Q-Q) plot is a graphical tool used in statistics to assess whether a dataset follows a particular theoretical distribution, often the normal distribution (a bell-shaped curve). It does this by comparing the quantiles (percentiles) of your dataset to the quantiles of the theoretical distribution you want to test against.<br>

Here's how a Q-Q plot works:<br>

1) **Data Preparation**: You start with a dataset that you want to test for conformity to a theoretical distribution. For example, you might have data on people's heights and want to check if it follows a normal distribution.<br>

2) **Sorting Data**: You sort the data in ascending order, from the smallest value to the largest.<br>

3) **Theoretical Quantiles**: You calculate the theoretical quantiles for the distribution you want to test. For a normal distribution, these quantiles are determined by the mean and standard deviation.<br>

4) **Plotting**: On a Q-Q plot, you place the observed data quantiles (from your dataset) on the y-axis and the corresponding theoretical quantiles (from the chosen distribution) on the x-axis.<br>

5) **Perfect Normality**: If your data perfectly follows the chosen distribution (e.g., normal distribution), the points on the Q-Q plot should fall along a straight diagonal line.<br>

6) **Deviation from Normality**: Any deviation from this diagonal line suggests that your data does not conform to the theoretical distribution. The direction and curvature of the points on the Q-Q plot can provide insights into the nature of this deviation. For example, points curving upward or downward can indicate skewness or heavy tails.<br>

In summary, a Q-Q plot is a visual tool to assess the fit of your data to a theoretical distribution. It's particularly useful for quickly identifying departures from normality, which can inform your choice of statistical methods or guide data transformations.<br>

```python
import statsmodels.api as sm
import scipy.stats as stats

# Convert your age list to a NumPy array
age_array = np.array(age)

# Create a Q-Q plot for the age data
sm.qqplot(age_array, line='s')  # 's' for standardized line

# Add labels and a title
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.title('Q-Q Plot for Age Data')

plt.show()
```

Based on the Q-Q plot, we can see that our data is skewed and does not closely follow the line.
This means that the age does not seem to be normally distributed.<br>

To confirm this, We will need to perform further statistical analysis.  

## A Practical Guide to Utilizing Statistical Tests for Data Analysis
1) **Null Hypothesis (H0)**: The null hypothesis generally assumes that your data follows a specific distribution (e.g., normal distribution). For example, in the Shapiro-Wilk test, H0 assumes that the data follows a normal distribution.  

2) **Alternative Hypothesis (H1)**: The alternative hypothesis contradicts the null hypothesis. It suggests that your data does not follow the assumed distribution.  

3) **Significance Level (Alpha)**: Alpha (α) is the predetermined threshold that you choose to determine whether to reject the null hypothesis. It represents the probability of making a Type I error (incorrectly rejecting a true null hypothesis). A common choice is alpha = 0.05, which corresponds to a 5% significance level.  

4) **Test Statistic**: When you perform a statistical test, it calculates a test statistic based on your data.  

5) **Critical Value**: The critical value(s) is a threshold derived from the chosen significance level and the test's statistical properties. If the test statistic exceeds the critical value, you reject the null hypothesis; otherwise, you fail to reject it.  

6) **Conclusion**: If the test statistic is greater than the critical value, you conclude that your data does not follow the assumed distribution (reject H0). If the test statistic is less than or equal to the critical value, you fail to reject the null hypothesis, indicating that your data is consistent with the assumed distribution.  

You can adjust the significance level (alpha) to control the trade-off between Type I and Type II errors (false positives and false negatives). A lower alpha (e.g., 0.01) makes it harder to reject the null hypothesis, reducing the risk of Type I errors but increasing the risk of Type II errors. Conversely, a higher alpha (e.g., 0.10) makes it easier to reject the null hypothesis but increases the risk of Type I errors.

In summary, you can use alpha as a tool to make decisions about your data's distribution based on the results of statistical tests. Adjusting alpha allows you to set the threshold for what you consider statistically significant evidence against the null hypothesis.

Our alpha will be set at 0.05 which means our confidence interval will be 95%.

## Shapiro-Wilk Test
The Shapiro-Wilk test evaluates whether a dataset follows a normal distribution i.e. it tests for normality.  
The null hypothesis (H0) of the test is that the data is normally distributed.  
If the p-value is less than your chosen significance level (e.g., 0.05), you would reject the null hypothesis, indicating that the data significantly deviates from normality.

## Shapiro-Wilk Test for Age

```python
from scipy import stats

# Specify your desired significance level (alpha)
alpha = 0.05

print(f'Size of data set for Age: {len(age)}')

# Perform the Shapiro-Wilk test with the specified alpha
shapiro_statistic_age, shapiro_p_value_age = stats.shapiro(age)
print(f"Shapiro-Wilk test statistic for age: {shapiro_statistic_age}")
print(f"P-value for age: {shapiro_p_value_age}")
print(f"Alpha: {alpha}")

# Check if the data is normally distributed based on the specified alpha
if shapiro_p_value_age > alpha:
    print("Age appears to be normally distributed (fail to reject H0)")
else:
    print("Age does not appear to be normally distributed (reject H0)")
    
# You can also check if the data is normally distributed based on the specified critical values and the test statistic
# but you will need to use statistical tables
```

## Anderson-Darling Test
The Anderson-Darling test is similar to the Shapiro-Wilk test but provides a more detailed assessment of the departure from normality.  
It provides critical values for several significance levels and a statistic that quantifies the deviation.  
You can compare the statistic to the critical values to determine whether the data significantly deviates from normality.

## Anderson-Darling Test for Age

```python
# Perform the Anderson-Darling test with a specified alpha of 0.05
anderson_statistic_age, anderson_critical_values_age, anderson_significance_level_age = stats.anderson(age, dist='norm')
print(f"Anderson-Darling test statistic for age: {anderson_statistic_age}")
print(f"Critical values of age at different significance levels [15%, 10%, 5%, 2.5%, 1%]: {anderson_critical_values_age}")
print(f"Significance level used: {alpha}")

# Check if the data is normally distributed based on the specified alpha
if anderson_statistic_age < anderson_critical_values_age[2]:  # Using the 95% significance level
    print("Age appears to be normally distributed (fail to reject H0)")
else:
    print("Age does not appear to be normally distributed (reject H0)")
```

Based on the Shapiro-Wilk test and the Anderson-Darling test, we can conclude that age does not follow a normal distribution, since the null hypothesis was rejected for both of these tests.
