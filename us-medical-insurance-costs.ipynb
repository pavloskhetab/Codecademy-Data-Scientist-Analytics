## Age Distribution
To view the age distribution we shall create a histogram.<br>
We will also calculate the mean, median, and standard deviation to summarise the age distribution.

```python
import matplotlib.pyplot as plt
import numpy as np

max_age = max(age)
min_age = min(age)

"""
Calculate the number of bins for a histogram using the Freedman-Diaconis rule.

Parameters:
data (list or numpy array): The data for which to calculate the number of bins.

Returns:
int: The number of bins.
"""
def calculate_num_bins(data):
    data_range = max(data) - min(data)  # Calculate the data range
    iqr = np.percentile(data, 75) - np.percentile(data, 25)  # Calculate the interquartile range
    bin_width = 2 * iqr / (len(data) ** (1/3))  # Calculate the bin width using Freedman-Diaconis rule
    num_bins = int(data_range / bin_width)  # Calculate the number of bins
    
    return num_bins, bin_width # Return both num_bins and bin_width

# Calculate the number of bins using the function
num_bins, bin_width = calculate_num_bins(age)

# Create the histogram with the specified number of bins and range
plt.hist(age, bins=num_bins, range=(min_age, max_age), edgecolor='k', alpha=0.8)


# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')

plt.show()

# Convert max_age and min_age to strings before printing
print('Max Age: ' + str(max_age))
print('Min Age: ' + str(min_age))
```
